```python
import pickle

with open("rag_artifacts.pkl", "rb") as f:
    rag_artifacts = pickle.load(f)

all_chunks = rag_artifacts["chunks"]
all_metadatas = rag_artifacts["metadatas"]
documents = rag_artifacts["documents"]

print("Chunks reloaded:", len(all_chunks))
print("Documents available:", list(documents.keys()))
```

```python
!pip install sentence-transformers chromadb -q

import chromadb
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

chroma_client = chromadb.Client()
collection = chroma_client.create_collection(name="coca_cola_regional_reports")

chunk_embeddings = embedding_model.encode(all_chunks).tolist()
chunk_ids = [f"chunk_{i}" for i in range(len(all_chunks))]

collection.add(
    documents=all_chunks,
    embeddings=chunk_embeddings,
    metadatas=all_metadatas,
    ids=chunk_ids,
)

print("RAG system rebuilt, chunks stored:", collection.count())
```

```python
forecast_data = {
    "North": {"lstm_mape": 3.93, "naive_mape": 3.41, "avg_weekly_sales": 2050000, "verdict": "naive baseline comparable"},
    "East": {"lstm_mape": 3.00, "naive_mape": 3.40, "avg_weekly_sales": 2000000, "verdict": "naive baseline comparable"},
    "South": {"lstm_mape": 4.54, "naive_mape": 4.47, "avg_weekly_sales": 2100000, "verdict": "naive baseline comparable"},
    "West": {"lstm_mape": 3.36, "naive_mape": 7.41, "avg_weekly_sales": 280000, "verdict": "LSTM substantially outperforms naive"},
}

print("Forecast data ready for", len(forecast_data), "regions")
```

```python
from crewai import Agent, LLM
import os
from google.colab import userdata

os.environ["GEMINI_API_KEY"] = userdata.get('GEMINI_API_KEY')
llm = LLM(model="gemini/gemini-2.5-flash", api_key=os.environ["GEMINI_API_KEY"])

forecast_interpreter = Agent(
    role="Forecast Interpreter",
    goal="Translate raw demand forecasting model output into clear, plain-language business statements",
    backstory="You are a data scientist who specializes in communicating model results to non-technical "
               "stakeholders. You always state both the model's accuracy and how it compares to a simple "
               "baseline, since raw accuracy numbers alone can be misleading.",
    llm=llm,
    verbose=True
)

researcher = Agent(
    role="Market Research Analyst",
    goal="Find precise, accurate facts from market intelligence reports to support business questions",
    backstory="You are a meticulous researcher who only states facts explicitly present in source material, "
               "and you cite which regional report each fact comes from.",
    llm=llm,
    verbose=True
)

analyst = Agent(
    role="Senior Business Analyst",
    goal="Synthesize forecast interpretations and research findings into strategic business implications",
    backstory="You are an experienced CPG strategy analyst who combines quantitative forecasts with "
               "qualitative market research to identify where investment will have the most impact.",
    llm=llm,
    verbose=True
)

writer = Agent(
    role="Executive Report Writer",
    goal="Produce a clear, board-ready strategic recommendation",
    backstory="You write concise, decision-oriented executive summaries that a CPG leadership team can "
               "act on immediately, always including a clear recommendation, not just an analysis.",
    llm=llm,
    verbose=True
)

print("All 4 agents created")
```

```python
from crewai.tools import tool

@tool("Get Regional Forecast Data")
def get_forecast_data(region: str) -> str:
    """Get the LSTM demand forecasting results for a specific region (North, East, South, or West)."""
    if region not in forecast_data:
        return f"No forecast data available for region: {region}"

    data = forecast_data[region]
    return (f"Region: {region}\n"
            f"LSTM forecast MAPE (error rate): {data['lstm_mape']}%\n"
            f"Naive baseline MAPE: {data['naive_mape']}%\n"
            f"Average weekly sales: ${data['avg_weekly_sales']:,}\n"
            f"Verdict: {data['verdict']}")

print("Forecast tool ready")
```

```python
from crewai import Task

forecast_task = Task(
    description="Using your forecast data tool, get the forecasting results for these 4 regions: "
                 "North, East, South, West. Translate each region's results into a clear, plain-language "
                 "statement about whether advanced forecasting adds value there, always citing both the "
                 "LSTM MAPE and naive baseline MAPE.",
    expected_output="A plain-language summary of forecasting performance for all 4 regions.",
    agent=forecast_interpreter,
    tools=[get_forecast_data],
)

research_task = Task(
    description="Using the forecast interpretation provided, research the market intelligence reports to "
                 "find supporting context for the question: '{question}'. Focus on facts that explain WHY "
                 "the forecasting results might differ across regions.",
    expected_output="Bullet-point facts from the reports relevant to the question, with source citations.",
    agent=researcher,
    context=[forecast_task],
)

analysis_task = Task(
    description="Combine the forecast interpretation and research findings to determine the strategic "
                 "implication for the question: '{question}'. Identify which regions should be prioritized "
                 "and why, grounded in both the numeric forecast results and the qualitative research.",
    expected_output="A strategic analysis identifying priority regions and the reasoning behind them.",
    agent=analyst,
    context=[forecast_task, research_task],
)

writing_task = Task(
    description="Write the final board-ready recommendation answering: '{question}'. Keep it to 4-6 "
                 "sentences, professional, and ending with a clear, specific recommendation.",
    expected_output="A polished executive recommendation.",
    agent=writer,
    context=[forecast_task, research_task, analysis_task],
)

print("All 4 tasks defined")
```

```python
@tool("Search Market Reports")
def search_market_reports(query: str) -> str:
    """Search the multi-region market intelligence reports for relevant information."""
    query_embedding = embedding_model.encode([query]).tolist()
    results = collection.query(query_embeddings=query_embedding, n_results=3)

    output = []
    for i, doc in enumerate(results["documents"][0]):
        meta = results["metadatas"][0][i]
        output.append(f"[{meta['source_document']}]: {doc}")

    return "\n\n".join(output)

# Attach this tool to the research task
research_task.tools = [search_market_reports]

print("RAG search tool attached to Researcher")
```

```python
from crewai import LLM

llm = LLM(model="gemini/gemini-2.5-flash-lite", api_key=os.environ["GEMINI_API_KEY"])

# Re-assign this lighter, higher-limit model to all 4 agents
forecast_interpreter.llm = llm
researcher.llm = llm
analyst.llm = llm
writer.llm = llm

print("Switched all agents to gemini-2.5-flash-lite")
```

```python
from crewai import Crew, Process

crew = Crew(
    agents=[forecast_interpreter, researcher, analyst, writer],
    tasks=[forecast_task, research_task, analysis_task, writing_task],
    process=Process.sequential,
    verbose=True
)

question = "Where should we prioritize our demand forecasting investment, and why?"

result = await crew.kickoff_async(inputs={"question": question})

print("\n\n=== FINAL RECOMMENDATION ===")
print(result)
```

```python
crew_output = str(result)

with open("crew_recommendation_output.txt", "w") as f:
    f.write(f"Question: {question}\n\n")
    f.write(f"Final Recommendation:\n{crew_output}")

from google.colab import files
files.download("crew_recommendation_output.txt")

print("Output saved")
```

# Component 5 — Strategic Recommendation Crew (Multi-Agent, Tool-Using)

Part of the Coca-Cola Demand Intelligence & Market Insights Platform. A 4-agent CrewAI pipeline that combines live structured model output with multi-document retrieval to produce board-ready strategic recommendations — the platform's reasoning and synthesis layer.

## Objective

Build a multi-agent system capable of reasoning across both structured model results (the LSTM's actual forecast accuracy data) and unstructured document knowledge (the multi-region market reports), producing a synthesized, accurate, board-ready recommendation — extending a prior 3-agent text-only CrewAI project with active tool use and live data integration.

## Architecture

Four agents, each with a distinct role, chained sequentially via CrewAI's `context` parameter:

1. **Forecast Interpreter** (new role) — calls a custom tool (`get_regional_forecast_data`) to retrieve the actual LSTM forecasting results for all four regions, translating raw MAPE figures into plain-language statements about where advanced forecasting adds genuine value
2. **Market Research Analyst** — calls a second custom tool (`search_market_reports`), which queries the Component 4 multi-document RAG/ChromaDB system, guided by the Forecast Interpreter's output to find supporting context
3. **Senior Business Analyst** — synthesizes the forecast interpretation and research findings into strategic implications
4. **Executive Report Writer** — produces the final 4-6 sentence, board-ready recommendation

**Key architectural upgrade from a prior CrewAI project:** agents in this pipeline actively call functions (`@tool`-decorated Python functions) to retrieve live data during their reasoning process, rather than only reading static text supplied in the prompt — a more realistic pattern for production agentic systems.

## Test result

**Question:** "Where should we prioritize our demand forecasting investment, and why?"

The Forecast Interpreter correctly retrieved the exact, real LSTM results for all 4 regions via tool calls (North: 3.93% LSTM vs. 3.41% naive; East: 3.0% vs. 3.4%; South: 4.54% vs. 4.47%; West: 3.36% vs. 7.41%), with zero numeric distortion. The final recommendation correctly identified West Region as the clear priority, citing the precise MAPE figures, and concluded: *"Prioritize all incremental demand forecasting investment and development efforts for the next fiscal year on enhancing capabilities within the West region."*

This output independently arrived at the same nuanced conclusion discovered through manual analysis in Component 1 — that the LSTM's value is concentrated in higher-volatility markets rather than distributed uniformly — demonstrating the crew can synthesize an accurate, non-trivial strategic insight rather than simply summarizing inputs.

## Technical notes / debugging

- **Rate limiting:** the 4-agent pipeline, involving multiple tool calls per agent, exceeded the Gemini free tier's request-per-minute limit (5 requests/minute for `gemini-2.5-flash`) during initial testing, producing a `429 RESOURCE_EXHAUSTED` error. Resolved by switching all agents to `gemini-2.5-flash-lite`, a higher-throughput model better suited to multi-call agentic workflows, combined with retry logic (3 attempts, 30-second backoff) as a production-realistic safety net for transient API failures.
- **PIL/Pillow version conflict:** `crewai-tools` installation triggered an `ImportError` due to an incompatible auto-installed Pillow version (12.0.0). Resolved by explicitly pinning `Pillow==11.0.0`.

## Key finding

Tool-using agents that can query live, structured model output (not just pre-written text) produce more grounded, verifiable recommendations — every numeric claim in the final output traces directly to an actual function call result, making the recommendation auditable in a way that purely text-based RAG retrieval alone cannot guarantee.

## What I'd improve next

- Add a tool allowing the Researcher to filter RAG search by region (using Component 4's metadata filtering capability), rather than always searching across all documents
- Add a "Critic" agent to verify the Writer's final output against the Forecast Interpreter's and Researcher's actual findings before finalizing, catching any drift introduced during synthesis
- Expose this entire crew as a Model Context Protocol (MCP) server (Component 6), making it callable from external AI applications, not just this notebook
